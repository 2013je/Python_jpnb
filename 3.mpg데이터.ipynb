```python
import numpy as np
mpg['cty_prepro']=np.where(mpg.loc[:,'cty'] >=17,1,0)
mpg
```

```python
# np.where(조건식): 해당 조건에 맞는 데이터의 행 정보(인덱스 정보) 출력
# 1. np.where(조건식,True값, False값)
# 2. 불리언 인덱싱 + 데이터 타입 변경하는 방법
mpg=mpg.assign(cty_prepro=np.where(mpg.loc[:,'cty'] >=17,1,0))
mpg
```

```python
cty_bool = mpg['cty']>=17
mpg['cty_prepro'] = cty_bool.astype('int32')
mpg
```

```python
mpg['cty_prepro'] = mpg['cty'].apply(lambda x: 1 if x >= 17 else 0)
# lambda 변수: 수행될 기능: 일시적으로 함수 생성(일회용)
```

```python
mpg=mpg.drop('cty_prepro',axis=1)
mpg.drop(['cty_prepro'])
# []대괄호를 통해 여러개 데이터명 넣어주고 한꺼번에 삭제 가능

mpg.drop('cty_prepro',axis=1, inplace=True) # 초기화를 True로 바꿈
```

```python
mpg.groupby(by='category').agg(cty_mean=('cty','mean'))
```

```python
mpg.groupby(by='category').agg(cty_mean=('cty','mean')).sort_values(by='cty_mean',ascending=False)
```

```python
mpg.groupby(by='manufacturer').agg(hwy_mean=('hwy','mean')).sort_values(by = 'hwy_mean',ascending=False).head(3)
```

```python
mpg.groupby(by='manufacturer').agg(compact_count=('category', 'count')).sort_values(by='compact_count', ascending=False)
mpg.groupby(by='manufacturer').agg(compact_count=('category', lambda x: (x == 'compact').sum()))
```
