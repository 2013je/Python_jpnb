```python
exam = pd.read_csv('./data/exam.csv')
def data_check(data):
    data_mean = data['math':].mean() # 한 사람에 대한 평균 출력
    # 85점 이상 A, 70점 이상 B, 60점 이상 C , 나머지 F
    #조건문
    if (data_mean>=85):
        return 'A'
    elif (data_mean>=70):
        return 'B'
    elif(data_mean>=60):
        return 'C'
    else:
        return 'F'

# 행방향 - 열 단위 처리(axis=0) 한 컬럼에 대한 정보들 확인
# 열방향 - 행 단위 처리(axis=1) 한 사람에 대한 정보들 확인
exam['grade']= exam.apply(data_check, axis = 1)
exam
```

- 카테고리 형식(범주형)으로 데이터 정리하기 - cut()



```python
ages = [0,2,15,21,23,37,31,61,20,41,32,100]
df_age=pd.DataFrame(ages, columns = ['ages'])
df_age

# 구간 범위 설정(범주 구간 설정)
# 구간 이름 설정('미성년자','청년','중년','장년','노년')
bins = [-1,15,25,35,60,100] # (-1,15],(15,25],(25,35],(35,60](60,100]
bin_label = ['미성년자','청년','중년','장년','노년']
cats = pd.cut(df_age['ages'],bins = bins, labels=bin_label)
df_age['category'] = cats # pd.cut(df_age['ages'],bins = bins, labels=bin_label )
df_age
```

```python
# category 컬럼 안에 있는 데이터 카운트
np.unique(df_age['category'], return_counts=True) 
df_age['category'].value_counts() # 결과물 series로 출력
# 1. 각 카테고리 묶어줌(그룹) 2. 해당 그룹에 묶인 데이터개수 몇개 cnt
```
