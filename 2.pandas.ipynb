- 1차원 Series 속성 확인

```python
# 값
num_sr2.values # numpy arr
# 인덱스
num_sr2.index # object : pandas 문자열 등 의미를 포함한 dtype
# 데이터 타입
num_sr2.dtype # int64 : 숫자를 표현할 수 있는 범위값 = 2**64
```

```python
population = pd.Series({'서울':9602000,'부산':334400,'광주':148800,'대구':2419000})
population.name='2020 인구수'
population.index.name='도시'
population
```

```python
population= pd.Series([9602000,3344000,1488000,2419000],
                      index=['서울','부산','광주','대구'])
population.name='2020 인구수'
population.index.name='도시'
```

- 1차원 시리즈 데이터 갱신(수정),추가

```python
# 갱신
# 부산>3500000
population['부산']=3500000
# 추가
population['대전']=1500000
```

```python
# 삭제
# population=population.drop('서울')
population.drop('서울',inplace=False) # 삭제해준 결과를 보여줌, 바로 초기화 하진 X
population.drop('서울',inplace=True) # inplace = True 삭제해준 결과를 반영
population
```

- 2차원 데이터프레임(DataFrame)생성1 - 단일 리스트에 값 입력하여 생성

```python
num = [1,1,3] # 단일리스트
num_df = pd.DataFrame(num) #일차원을 넣으면
num_df.ndim # 이차원이됨
```

- 2차원 데이터프레임 생성 2 - 이중리스트에 값 입력하여 생성

```python
num2 = [['iot',1],['big',1],['ai',3]]
num2_df=pd.DataFrame(num2,columns=['Class','Join'],index=[1,2,3])
num2_df
# index 값을 설정하려면 사용하는 키워드 : index
```

- 2차원 데이터프레임 생성2 - 딕셔너리로 데이터 입력하여 생성

```python
num2_df = pd.DataFrame({'Class':['iot','big','ai'],'Join':[1,1,3]},index=range(1,4))
num2_df
# 1차원 시리즈 생성 시 키 값은 인덱스명
# 2차원 데이터프레임 생성시 키 값은 컬럼명
# index 값 설정하고 싶으면, index = [인덱스명, 인덱스명, ...]
```

```python
# 2차원 데이터프레임 속성확인
# 벨류 () 확인
display(num2_df.values)
# 인덱스 확인
display(num2_df.index)
# 컬럼명 확인
display(num2_df.columns)
# 데이터 타입
display(num2_df.dtypes)
```

#### pandas 데이터 접근

- 시리즈 데이터 접근 1 - 인덱스 번호로 접근

```python
# 경고 메세지 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
num_sr2[0]
num_sr2[1]
num_sr2[2]
```

- 1차원 시리즈 데이터 접근 - 인덱스 값으로 접근

```python
num_sr2['son']
num_sr2['kim']
num_sr2['park']
```

- 1차원 시리즈 데이터 접근 - 여러개 접근

```python
num_sr2[['son','park']]

# 김과 박 인덱스 번호 활용해서 접근
num_sr2[[1,2]]

# 슬라이싱(범위 지정) > 끝값 미포함
num_sr2[1:3]

# 문자열로 슬라이싱 > 끝값 포함!
num_sr2['kim':'park']
num_sr2['kim':]
```

- 2차원 데이터프레임 열(컬럼) 접근

```python
# df[열값]
num2_df['Class']

# Join 컬럼 접근
num2_df['Join']

# 컬럼 여러개 접근
num2_df[['Join',]]
```

- 데이터 접근 인덱서
    - df.loc[행,열] 인덱스 값(눈에 보이는)으로 접근하는 인덱서
    - df.iloc[행,열] 인덱스 번호(숫자)로 접근하는 인덱서

```python
# numpy 2차원 데이터 접근
# arr2[행,열]
# arr2[행:행, 열:열]
```

```python
# Class 컬럼을 인덱스 값으로 바꾸기(설정하기)
# num2_df = num2_df.set_index('Class')
num2_df

# 인덱스 번호로 접근 
num2_df.iloc[0]

# 인덱스 값으로 접근
num2_df.loc['iot']
```

```python
# loc, iloc 사용> 'ai'행 접근
num2_df.loc['ai']
num2_df.iloc[2]
```

```python
#
num2_df.iloc[[0,2]]
display(num2_df.loc[['iot','big']])
display(num2_df.loc['iot':'big'])
```

- 2차원 데이터 프레임- 인덱서로 열 접근

```python
# 0번째 열
display(num2_df.iloc[:,0])
# 'Join' 열 접근
display(num2_df.loc[:,'Join':]) # 시리즈가 끝이 더 있는 줄 알고 2차원으로 인식
display(num2_df.loc[:,['Join']])
```

```python
person_info
# 연재의 성별은 무엇인가?
display(person_info.loc[1,'성별'])
display(person_info.iloc[1,2])
```

```python
display(person_info.loc[0,'이름':'나이'])
display(person_info.loc[2,'이름':'나이'])
```

```python
person_info.iloc[[0,2],[0,1]]
person_info.iloc[[0,2],0:2] # [[0,2],0:2:1]
person_info.iloc[::2,:2] # ::2 처음부터 끝까지 2칸 띄면서 데이터 접근
```
