## 시각화 도구 사용하기

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 추가적으로 시각화 하는 도구
import seaborn as sns
```

### Line plot 선그래프

```python
# 배열 데이터 생성하기
x= np.arange(1,6)
# seed 랜덤한 수를 고정하는 값
np.random.seed(3)
y =np.random.randint(1,10,size=5)
print(x,y)
```

```python
# 선그래프 그리기
plt.plot(x,y, c='red', lw=3, ls=':', marker='s', ms = 9 , mec='blue',mew=3,mfc='green')
plt.show()
```

```python
plt.plot(x,y, marker='*', mew = 3, mec = 'orange',mfc = 'yellow', ms = 15, c = 'c')
plt.show()
```

```python
x=[1,2,3]
y1=[1,2,3]
y2=np.array([4,5,3])
y3=np.array([3,6,2])
plt.plot(x,y1,label='y1')
plt.plot(x,y2,label='y2')
plt.plot(x,y3,label='y3')
plt.legend()#범례 표시
plt.show()
plt.plot(x,y3,label='y3') # 범례 표시 X(구분 지어줬기 때문)
plt.show()
```

- subplot() - 한창에 칸 여러개를 나눠서 여러개 동시 표현

```python
# x, y1, y2, y3
# 창(도화지) 크기 설정
plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.title('<1>')
plt.plot(x,y1)

plt.subplot(1,3,2)
plt.title('<2>')
plt.plot(x,y2)

plt.subplot(1,3,3)
plt.title('<3>')
plt.plot(x,y3)
plt.show()
```

```python
data=pd.read_csv('./data/장래_인구변동_KOSIS.csv', 
                 encoding='EUC-KR', index_col='인구변동요인별')
data
# data.info()
# data.shape
# data.index
```

### Box plot 상자그래프
 - 수치형태의 데이터를 표현할 때 자주 사용 그래프

```python
# 1d 데이터 준비 > 박스 플롯 그리기
tmp_data=np.random.randint(1,10,size=10)
plt.boxplot(tmp_data, labels=['데이터명'])
plt.show()
```

```python
# 2d 데이터 준비 > 박스 플롯 그리기
tmp_data=np.random.randint(1,49,size=(10,2))
tmp_data
plt.boxplot(tmp_data,labels=['1열','2열'])
plt.show()
# 2차원 데이터 > box plot 2개 그려짐
# 컬럼(열) 한개씩 인식 > 그래프 그려줌
```

 입력 받은 월의 평균기온 구하기


```python
td.iloc[:,0]
td['월']=td['날짜'].str[6:8] # 파생변수
month = input("월 입력(01~12)>> ")
td_mon=td.query(f"월=='{month}'")

plt.boxplot(td_mon['평균기온(℃)'],labels=[f'{month}월 평균'])
plt.show()
```
