## 시각화 도구 사용하기

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 추가적으로 시각화 하는 도구
import seaborn as sns
```

### Line plot 선그래프

```python
# 배열 데이터 생성하기
x= np.arange(1,6)
# seed 랜덤한 수를 고정하는 값
np.random.seed(3)
y =np.random.randint(1,10,size=5)
print(x,y)
```

```python
# 선그래프 그리기
plt.plot(x,y, c='red', lw=3, ls=':', marker='s', ms = 9 , mec='blue',mew=3,mfc='green')
plt.show()
```

```python
plt.plot(x,y, marker='*', mew = 3, mec = 'orange',mfc = 'yellow', ms = 15, c = 'c')
plt.show()
```

```python
x=[1,2,3]
y1=[1,2,3]
y2=np.array([4,5,3])
y3=np.array([3,6,2])
plt.plot(x,y1,label='y1')
plt.plot(x,y2,label='y2')
plt.plot(x,y3,label='y3')
plt.legend()#범례 표시
plt.show()
plt.plot(x,y3,label='y3') # 범례 표시 X(구분 지어줬기 때문)
plt.show()
```

- subplot() - 한창에 칸 여러개를 나눠서 여러개 동시 표현

```python
# x, y1, y2, y3
# 창(도화지) 크기 설정
plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.title('<1>')
plt.plot(x,y1)

plt.subplot(1,3,2)
plt.title('<2>')
plt.plot(x,y2)

plt.subplot(1,3,3)
plt.title('<3>')
plt.plot(x,y3)
plt.show()
```

### 예제1
 - 인구데이터 분석 실습 - 장래 인구변동 데이터를 이용하여 lineplot로 출생아수, 사망자수

```python
data=pd.read_csv('./data/장래_인구변동_KOSIS.csv', 
                 encoding='EUC-KR', index_col='인구변동요인별')
data
# data.info()
# data.shape
# data.index
```

```python
# 그래프에 한글 보이기
# 1. 한글 폰트 경로
# 2. 폰트 이름 확인
# 3. rc 도구에 폰트 이름 연결
from matplotlib import font_manager, rc
path='C:/Windows/Fonts/malgun.ttf'
font_name= font_manager.FontProperties(fname=path).get_name()
rc('font',family=font_name)
```

```python
# 요약 가능
from matplotlib import font_manager, rc
rc('font',family='malgun')
```

```python
x = range(2020,2071,10)
y1= data.loc['출생아수(천명)']
y2= data.loc['사망자수(천명)']

plt.figure(figsize=(6,4))
plt.title("장래 인구변동 시각화") # 한글 폰트 지원하도록 연결
plt.plot(x,y1,label='출생아수(천명)',marker='*',ms=10)
plt.plot(x,y2,label='사망자수(천명)',marker='o')

plt.xlabel('년도'); plt.ylabel('인구수')
plt.xticks(range(2020,2071,5))
# plt.ylim(500,750) # 줌인
plt.legend(); plt.grid()
# plt.show()
plt.savefig('./data/kosis 장래 인구변동 시각화2.png',dpi=600,bbox_inches='tight')
```
